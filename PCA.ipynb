```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
pose_region = pd.read_csv('pose_region_props.csv')
poselandmark = pd.read_csv('poselandmarks_props.csv')
```

```python
pose_region.sample(5)
```

```python
pose_region.describe()
```

```python
#mean centering of data
mean_scaling = StandardScaler()
pose_region_scaled = pd.DataFrame(mean_scaling.fit_transform(pose_region.iloc[:,:-1]), columns= pose_region.columns[:-1])
```

```python
pose_region_scaled.sample(5)
```

```python
eigen_values, eigen_vec = np.linalg.eig(np.cov(pose_region_scaled,rowvar=False))
```

```python
eigen_values
```

```python
pca = PCA(n_components=2)
pc_comps = pd.DataFrame(pca.fit_transform(pose_region_scaled))
```

```python
import plotly.express as px
```

```python
labels = pose_region.iloc[:,-1]
figure = px.scatter(x=pc_comps.iloc[:,0],
                    y=pc_comps.iloc[:,1],
                    color=labels,
                    color_discrete_sequence=px.colors.qualitative.G10)

figure.show()
```

```python
pca = PCA(n_components=3)
feature_3_pca = pd.DataFrame(pca.fit_transform(pose_region_scaled), columns=['f1','f2','f3'])
```

```python
feature_3_pca.sample(5)
```

```python
labels = pose_region.iloc[:,-1]
figure = px.scatter_3d(feature_3_pca, x='f1', y='f2', z='f3',
                        color=labels)
                        

figure.show()
```
